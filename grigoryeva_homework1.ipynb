```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

def compute_polynomial_coeffs(n):
    '''
    n - количество точек
    '''
    xx = np.linspace(1, 15, n)
    yy = f(xx)
    b = yy
    A = np.zeros((n, n))
    for i in range(n):
        A[:,i] = xx**i
    return np.linalg.solve(A, b)

def polynome(w, x):
    n = len(w)
    result = 0
    for i in range(n):
        result += w[i] * x**i
    return result
```

```python
n = 5

w = compute_polynomial_coeffs(n)
xx = np.linspace(1, 15, 100)
plt.plot(xx, f(xx), label='f(x)', color = 'purple')
plt.plot(xx, polynome(w, xx), '--', label='polynome', color = 'green')

x_points = np.linspace(1, 15, n)
y_points = f(x_points)
plt.plot(x_points, y_points, 'o', label='points', color = 'red')
plt.legend()
plt.grid()
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

def mnk(xx, yy):
    '''
    xx, yy - набор точек и значений неизвестной функции
    '''
    N = len(xx)
    b = yy
    A = np.ones([N, n])
    for i in range(1, n):
        A[:, i] = xx ** i
    return np.linalg.inv(A.T.dot(A)).dot(A.T).dot(b)

def polynome(w, x):
    N = len(w)
    result = 0
    for i in range(N):
        result += w[i] * x**i
    return result
```

```python
N = 100
n = 5
noise = 0.1
x_points = np.linspace(1, 15, N)
y_points = f(x_points) + np.random.randn(N) * noise

w = mnk(x_points, y_points)

xx = np.linspace(1, 15, 100)
plt.plot(x_points, f(x_points), label='f(x)', color = 'purple')
plt.plot(x_points, polynome(w, x_points), '--', label='polynome', color = 'green')
plt.plot(x_points, y_points, 'o', label='points', color = 'red', markersize = 3)
plt.legend()
plt.grid()
```
